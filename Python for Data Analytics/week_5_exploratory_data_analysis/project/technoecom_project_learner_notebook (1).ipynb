```python
def perform_bivariate_analysis(data, column ):
    if data[column].dtype == 'object':
        # Categorical variable

        plt.figure(figsize=(15,7))
        sns.swarmplot(data=data,x=column,y=x_column)
        plt.show()

        result = data.groupby(column)[x_column].mean()
        print(f"Bivariate analysis between {x_column} and '{column}':")
        print(result)

    else:
        # Integer variable
        print(f'Title: Bivariate Numerical column - Scatterplot of {column}')
        plt.figure(figsize=(10,6))
        sns.scatterplot(data=data,x=x_column,y=column)
        plt.show()

        result = data.groupby(pd.cut(data[column], bins=5))[x_column].mean()
        print(f"Bivariate analysis between {x_column} and '{column}':")
        print(result)
```
